```python
import matplotlib.pyplot as plt

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Function to convert Fahrenheit to Celsius
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

# Generate data for Celsius and corresponding Fahrenheit values
celsius_values = list(range(-30, 51))  # From -30째C to 50째C
fahrenheit_values_from_celsius = [celsius_to_fahrenheit(c) for c in celsius_values]

# Generate data for Fahrenheit and corresponding Celsius values
fahrenheit_values = list(range(-22, 123))  # From -22째F to 122째F
celsius_values_from_fahrenheit = [fahrenheit_to_celsius(f) for f in fahrenheit_values]

# Debugging: Print lengths of lists
print("Length of celsius_values:", len(celsius_values))
print("Length of fahrenheit_values_from_celsius:", len(fahrenheit_values_from_celsius))
print("Length of fahrenheit_values:", len(fahrenheit_values))
print("Length of celsius_values_from_fahrenheit:", len(celsius_values_from_fahrenheit))

# Debugging: Check for mismatched lengths
if len(celsius_values) != len(fahrenheit_values_from_celsius):
    print("Mismatch: celsius_values and fahrenheit_values_from_celsius have different lengths!")
if len(fahrenheit_values) != len(celsius_values_from_fahrenheit):
    print("Mismatch: fahrenheit_values and celsius_values_from_fahrenheit have different lengths!")

# Create the plot
plt.figure(figsize=(10, 6))

# Plot Celsius to Fahrenheit conversion
plt.plot(celsius_values, fahrenheit_values_from_celsius, label="Celsius to Fahrenheit", color="blue")

# Plot Fahrenheit to Celsius conversion
plt.plot(celsius_values_from_fahrenheit, fahrenheit_values, label="Fahrenheit to Celsius", color="red")

# Add labels and title
plt.xlabel("Temperature")
plt.ylabel("Converted Temperature")
plt.title("Temperature Conversion: Celsius to Fahrenheit and Vice Versa")

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
