```python
import matplotlib.pyplot as plt


grocery_list = {'apples': 5, 'bananas': 3, 'carrots': 4, 'mango': 2, 'kiwi': 1, 'cherry': 6, 'orange': 9}


items = list(grocery_list.keys())
quantities = list(grocery_list.values())


plt.figure(figsize=(8, 8))
plt.pie(quantities, labels=items, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'pink'})


centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Items', labels=items)

plt.title('Grocery List')
plt.axis('equal')
plt.show()
```
