```python
import matplotlib.pyplot as plt


grocery_list = {'apples': 5, 'bananas': 3, 'carrots': 4,'mango':2,'kiwi':1,'cherry':6,'orange':9}


items = list(grocery_list.keys())
quantities = list(grocery_list.values())


plt.figure(figsize=(8, 8))
plt.pie(quantities, labels=items, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'pink'})


centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.title('Grocery List')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grocery list with quantities
grocery_list = {'apples': 5, 'bananas': 3, 'carrots': 4, 'mango': 2, 'kiwi': 1, 'cherry': 6, 'orange': 9}

items = list(grocery_list.keys())
quantities = list(grocery_list.values())

# Create a figure with two subplots: one for the pie chart and one for the color bar
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the pie chart on the left subplot
wedges, texts, autotexts = ax1.pie(quantities, labels=items, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'pink'})
ax1.set_title('Grocery List')

# Create a color bar on the right subplot
color_palette = plt.cm.get_cmap('tab20c', len(items))  # Choose a color map and number of colors
colors = [color_palette(i) for i in range(len(items))]
ax2.barh(items, quantities, color=colors)
ax2.set_xlabel('Quantities')
ax2.set_ylabel('Items')
ax2.set_title('Quantities')

plt.tight_layout()
plt.show()
```
